```python
import csv
import cv2 
import numpy as np

lines = []

with open('../Desktop/sdc_bc_training/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        lines.append(line)
        
        
images = []
measurements = []
pathlist = []

for line in lines:
    source_path = line[0]
    filename = source_path.split('/')[-1]
    current_path = '/Users/christophersheehan/Desktop/sdc_bc_training/IMG' + '/' + filename
    
    image_centre = cv2.imread(current_path)
    
    source_path = line[1]
    filename = source_path.split('/')[-1]
    current_path = '/Users/christophersheehan/Desktop/sdc_bc_training/IMG' + '/' + filename
    
    pathlist.append(current_path)
    
    image_left = cv2.imread(current_path)
    
    source_path = line[2]
    filename = source_path.split('/')[-1]
    current_path = '/Users/christophersheehan/Desktop/sdc_bc_training/IMG' + '/' + filename
    
    image_right = cv2.imread(current_path)

    image_full = np.vstack( (image_left, image_centre, image_left) )
    
    print(image_full.shape)
    images.append(image_full)
    measurement = float(line[3])
    measurements.append(measurement)
    


X_train = np.array(images)
y_train = np.array(measurements)

```

```python

from keras.models import Sequential
from keras.layers import Flatten, Dense, Activation, Convolution2D

model = Sequential()



model.add(Flatten(input_shape=(480, 320, 3) ))
model.add(Dense(12))
model.add(Activation('relu'))
model.add(Dense(12))
model.add(Activation('relu'))
model.add(Dense(12))
model.add(Activation('relu'))
model.add(Dense(1))


model.compile(loss='mse', optimizer='adam')

model.fit(X_train, y_train, validation_split=0.2, shuffle=True,
         nb_epoch=5)
model.save('model.v4')




```
